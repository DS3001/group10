```python
import pandas as pd
import numpy as np
df_train = pd.read_csv('./data/training_data.csv')
df_test = pd.read_csv('./data/testing_data.csv')
```

```python
y_train = df_train['stroke']
X_train = df_train.drop(columns = ['stroke', 'Unnamed: 0', 'id'],axis=1)
y_test = df_test['stroke']
X_test = df_test.drop(columns = ['stroke', 'Unnamed: 0', 'id'],axis=1)
```

```python
X_train['bmi'] = X_train['bmi'].fillna(X_train['bmi'].mean())
X_test['bmi'] = X_test['bmi'].fillna(X_test['bmi'].mean())
```

```python
df_train
```

```python
# I used the same mapping as Olivia and Autumn, so we could combine models later
X_train['ever_married'] = X_train['ever_married'].replace({"No":0, "Yes":1})
X_test['ever_married'] = X_test['ever_married'].replace({"No":0, "Yes":1})

X_train['gender'] = X_train['gender'].replace({"Male":0, "Female":1, "Other":2})
X_test['gender'] = X_test['gender'].replace({"Male":0, "Female":1, "Other":2})

X_train['smoking_status'] = X_train['smoking_status'].replace({"never smoked":0, "formerly smoked":1, "smokes":2, 
                                                          "Unknown":3})
X_test['smoking_status'] = X_test['smoking_status'].replace({"never smoked":0, "formerly smoked":1, "smokes":2, 
                                                          "Unknown":3})

X_train['Residence_type'] = X_train['Residence_type'].replace({"Urban":0, "Rural":1})
X_test['Residence_type'] = X_test['Residence_type'].replace({"Urban":0, "Rural":1})
```

```python
# added a work_type mapping
X_train["work_type"].value_counts()
```

```python
X_train['work_type'] = X_train['work_type'].replace({"Private":0, "Self-employed":1, "children":2, 
                                                          "Govt_job":3, "Never_worked":4})
X_test['work_type'] = X_test['work_type'].replace({"Private":0, "Self-employed":1, "children":2, 
                                                          "Govt_job":3, "Never_worked":4})
```

```python
X_train
```

```python
# Now with all the data in numerical format, I see that avg_gluclose_level is sometimes over 100 so compared to the other 
# variables. However, I think it is the most important variable, so I will build two models, one with this data and another
# using the inver hyperbolic sine og avg_glucose_level
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.preprocessing import PolynomialFeatures
```

```python
reg = LinearRegression().fit(X_train, y_train) # Fit the linear model
print("Coefficents", reg.coef_)
print('R2: ', reg.score(X_test, y_test)) # R squared measure
y_hat = reg.predict(X_test)
N = len(y_test)
print('RMSE: ', (np.sum( (y_test - y_hat)**2)/N )**.5 )
```

```python
X_train_numeric = X_train.loc[:,['age','hypertension','heart_disease','avg_glucose_level','avg_glucose_level_ihs', 'smoking_status'] ]

expander = PolynomialFeatures(degree=1,include_bias=False) # Create the expander
Z = expander.fit_transform(X_train_numeric) # Pass the df into the expander to get powers/interactions of x and y
continuous = pd.DataFrame(data=Z) # Create a new, expanded dataframe

dummies = pd.concat([ pd.get_dummies(X_train['work_type'],dtype='int',drop_first=True),
                      pd.get_dummies(X_train['Residence_type'],dtype='int',drop_first=True),
                      #pd.get_dummies(X_train['smoking_status'],dtype='int',drop_first=True),
                    pd.get_dummies(X_train['ever_married'],dtype='int',drop_first=True)],axis=1)

X_expanded_train = pd.concat([continuous,dummies],axis=1)
```

```python
X_test_numeric = X_test.loc[:,['age','hypertension','heart_disease','avg_glucose_level',
                               'avg_glucose_level_ihs', 'smoking_status'] ]

expander = PolynomialFeatures(degree=1,include_bias=False) # Create the expander
Z = expander.fit_transform(X_test_numeric) # Pass the df into the expander to get powers/interactions of x and y
continuous = pd.DataFrame(data=Z) # Create a new, expanded dataframe

dummies = pd.concat([ pd.get_dummies(X_test['work_type'],dtype='int',drop_first=True),
                      pd.get_dummies(X_test['Residence_type'],dtype='int',drop_first=True),
                      #pd.get_dummies(X_test['smoking_status'],dtype='int',drop_first=True),
                    pd.get_dummies(X_test['ever_married'],dtype='int',drop_first=True)],axis=1)

X_expanded_test = pd.concat([continuous,dummies],axis=1)
```

```python
reg = LinearRegression().fit(X_expanded_train, y_train) # Fit the linear model
print("Coefficents", reg.coef_)
print('R2: ', reg.score(X_expanded_test, y_test)) # R squared measure
y_hat = reg.predict(X_expanded_test)
N = len(y_test)
print('RMSE: ', (np.sum( (y_test - y_hat)**2)/N )**.5 )
```
